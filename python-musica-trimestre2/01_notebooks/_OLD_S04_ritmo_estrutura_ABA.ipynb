## 1) Lembranza: Stream + comp√°s + tempo

Vamos crear un `Stream` e un contexto musical com√∫n:
- 4/4
- tempo = 100
- tonalidade: Do maior


```python
def crear_base(time_sig="4/4", bpm=100, tonalidade="C"):
    s = stream.Stream()
    s.append(meter.TimeSignature(time_sig))
    s.append(tempo.MetronomeMark(number=bpm))
    s.append(key.Key(tonalidade))
    return s

base = crear_base()
base.show("text")
```

## 2) Representar unha frase como lista de eventos

Hoxe traballaremos as√≠:
- `alturas` = lista de notas (strings)
- `ritmo` = lista de duraci√≥ns (en pulsos: negra=1)

Logo converteremos iso a notas de `music21`.


```python
alturas_A = ["C4","D4","E4","G4","E4","D4","C4","E4"]
ritmo_A   = [1,1,1,1,  1,1,1,1]  # 2 compases en 4/4

list(zip(alturas_A, ritmo_A))
```

### Conversor: listas -> Stream

Imos crear unha funci√≥n que:
- reciba alturas e duraci√≥ns
- devolva un `Stream` coas notas


```python
def frase_a_stream(alturas, ritmos, time_sig="4/4", bpm=100, tonalidade="C"):
    s = crear_base(time_sig=time_sig, bpm=bpm, tonalidade=tonalidade)
    for p, d in zip(alturas, ritmos):
        s.append(note.Note(p, quarterLength=d))
    return s

A = frase_a_stream(alturas_A, ritmo_A, bpm=100, tonalidade="C")
A.show("text")
```

## 3) Variaci√≥n r√≠tmica mantendo alturas

A idea:
- mesmas `alturas_A`
- un novo `ritmo_A2` distinto, pero que siga cadrando en compases.

Exemplo: se A dura 2 compases (8 pulsos), A2 tam√©n debe sumar 8.


```python
def suma_pulsos(ritmos):
    return sum(ritmos)

ritmo_A2 = [0.5,0.5,1,  0.5,0.5,1,  2,2]  # suma 8
print("Suma:", suma_pulsos(ritmo_A2))

A2 = frase_a_stream(alturas_A, ritmo_A2, bpm=100, tonalidade="C")
A2.show("text")
```

### Exercicio r√°pido
Crea un `ritmo_A3` que:
- sume 8 pulsos
- te√±a polo menos un silencio (Rest)

Pista: para meter un silencio, podes usar a altura `"REST"`.


```python
def frase_a_stream_con_rest(alturas, ritmos, time_sig="4/4", bpm=100, tonalidade="C"):
    s = crear_base(time_sig=time_sig, bpm=bpm, tonalidade=tonalidade)
    for p, d in zip(alturas, ritmos):
        if p == "REST":
            s.append(note.Rest(quarterLength=d))
        else:
            s.append(note.Note(p, quarterLength=d))
    return s

# TODO: propo√±er un ritmo e unha lista con REST
alturas_A3 = ["C4","D4","REST","G4","E4","D4","C4","E4"]
ritmo_A3   = [1,1,1,1,  1,1,1,1]  # suma 8
print("Suma:", suma_pulsos(ritmo_A3))

A3 = frase_a_stream_con_rest(alturas_A3, ritmo_A3)
A3.show("text")
```

## 4) Constru√≠r a secci√≥n B (contraste)

A secci√≥n B debe contrastar con A:
- outras alturas (p.ex. m√°is agudas)
- ou outro ritmo (m√°is longas / m√°is sincopado)
- ou outro contorno (sube moito / baixa moito)

Imos facer B tam√©n de 2 compases (8 pulsos).


```python
alturas_B = ["E4","F4","G4","A4","G4","F4","E4","D4"]
ritmo_B   = [1,1,1,1,1,1,1,1]  # 8 pulsos

B = frase_a_stream(alturas_B, ritmo_B, bpm=100, tonalidade="C")
B.show("text")
```

## 5) Montar a forma A‚ÄìB‚ÄìA

O truco: facer un novo `Stream` e ir engadindo as notas de cada secci√≥n.


```python
def unir_seccions(*seccions, time_sig="4/4", bpm=100, tonalidade="C"):
    s = crear_base(time_sig=time_sig, bpm=bpm, tonalidade=tonalidade)
    for sec in seccions:
        for el in sec.recurse().notesAndRests:
            s.append(el)
    return s

ABA = unir_seccions(A, B, A, bpm=100, tonalidade="C")
ABA.show("text")
```
