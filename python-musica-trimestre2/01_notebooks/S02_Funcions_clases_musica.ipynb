## 2) Funci√≥ns: xestos reutilizables


```python
def tocar_melodia(mel):
    for nota in mel:
        print(nota)

tocar_melodia(melodia)
```

### Exercicio guiado
- Crea unha funci√≥n `repetir(mel, veces)`


```python
def repetir(mel, veces):
    resultado = []
    for _ in range(veces):
        resultado += mel
    return resultado

print(repetir(melodia, 2))
```

## 4) Clases: notas como obxectos


```python
class Nota:
    def __init__(self, pitch, dur):
        self.pitch = pitch
        self.dur = dur

    def __repr__(self):
        return f"Nota({self.pitch}, {self.dur})"

frase = [
    Nota("C4", 1),
    Nota("D4", 1),
    Nota("E4", 2)
]

frase
```

### Transformaci√≥n sobre obxectos


```python
def alongar(frase, factor):
    return [Nota(n.pitch, n.dur * factor) for n in frase]

alongar(frase, 2)
```

## 5) Canon simple con desfase

Un canon √© unha repetici√≥n con atraso.


```python
def canon(frase, desfase):
    silencio = [Nota("REST", 1) for _ in range(desfase)]
    return frase + silencio + frase

canon(frase, 2)
```
