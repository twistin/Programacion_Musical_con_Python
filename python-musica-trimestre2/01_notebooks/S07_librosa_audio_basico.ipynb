```python
!pip -q install librosa soundfile

import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import soundfile as sf

from IPython.display import Audio, display

print("‚úÖ librosa listo")
```

## 1) Que √© `y` e que √© `sr`?

Cando cargamos audio obtemos:
- `y`: array de mostras (amplitudes no tempo)
- `sr`: sampling rate (mostras por segundo), por exemplo 22050 Hz

Duraci√≥n (en segundos) = `len(y) / sr`


## 2) Cargar un audio de exemplo

`librosa` trae exemplos (para non depender de subir ficheiros hoxe).
Logo, se queres, poder√°s subir un `.wav` do teu ordenador.


```python
# Audio de exemplo inclu√≠do en librosa
audio_path = librosa.ex("trumpet")  # tam√©n hai "choice", "brahms", etc.
y, sr = librosa.load(audio_path, sr=None)  # sr=None -> mant√©n o sampling rate orixinal

duracion = len(y) / sr
print("Ruta:", audio_path)
print("Sampling rate (sr):", sr)
print("Mostras:", len(y))
print("Duraci√≥n (s):", round(duracion, 2))
```

### Escoitar o audio en Colab


```python
display(Audio(y, rate=sr))
```

## 3) Visualizar a forma de onda

A forma de onda mostra:
- o tempo no eixo X
- a amplitude no eixo Y

Non √© ‚Äúaltura musical‚Äù, √© **amplitude** (enerx√≠a no tempo).


```python
plt.figure()
librosa.display.waveshow(y, sr=sr)
plt.title("Forma de onda (waveform)")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show()
```

## 4) Traballar con duraci√≥ns e fragmentos

Imos cortar:
- fragmento A: de 0s a 2s
- fragmento B: de 2s a 4s

‚ö†Ô∏è Importante: para cortar usamos √≠ndices en mostras:
- 1 segundo = `sr` mostras


```python
def cortar(y, sr, inicio_s, fin_s):
    i0 = int(inicio_s * sr)
    i1 = int(fin_s * sr)
    return y[i0:i1]

frag_A = cortar(y, sr, 0, 2)
frag_B = cortar(y, sr, 2, 4)

print("Dur A:", round(len(frag_A)/sr, 2), "s")
print("Dur B:", round(len(frag_B)/sr, 2), "s")

display(Audio(frag_A, rate=sr))
display(Audio(frag_B, rate=sr))
```

### Exercicio r√°pido

```python
# TODO: cambia os valores para o teu corte
inicio, fin = 0.5, 1.5
frag_C = cortar(y, sr, inicio, fin)

print("Dur C:", round(len(frag_C)/sr, 2), "s")
display(Audio(frag_C, rate=sr))
```

## 5) Comparar fragmentos visualmente

Imos debuxar 2 ondas para comparar:
- fragmento A
- fragmento B


```python
plt.figure()
librosa.display.waveshow(frag_A, sr=sr)
plt.title("Fragmento A (0‚Äì2s)")
plt.show()

plt.figure()
librosa.display.waveshow(frag_B, sr=sr)
plt.title("Fragmento B (2‚Äì4s)")
plt.show()
```

## 6) Gardar fragmentos en WAV

Usamos `soundfile` para gardar audio.
Despois podes descargar os `.wav` desde o panel de ficheiros de Colab.


```python
sf.write("frag_A.wav", frag_A, sr)
sf.write("frag_B.wav", frag_B, sr)
sf.write("frag_C.wav", frag_C, sr)

print("‚úÖ Gardados: frag_A.wav, frag_B.wav, frag_C.wav")
```
