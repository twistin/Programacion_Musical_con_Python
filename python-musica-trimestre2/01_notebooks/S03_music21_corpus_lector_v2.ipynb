## 1) Cargar unha obra do corpus
`music21.corpus.parse()` permite cargar repertorio (Bach, Beethoven, Mozart, etc.)

Exemplo cl√°sico para Harmon√≠a: **corais de Bach**.


```python
from music21 import corpus

score = corpus.parse("bach/bwv66.6")  # coral moi usado en exemplos
score
```

### Ver a partitura
En Colab √°s veces non se renderiza bonito, pero **podemos exportar** a MusicXML.


```python
# Vista textual (√∫til para entender estrutura)
score.show("text")
```

## 2) Estrutura: Score ‚Üí Parts ‚Üí Measures
- `Score`: obra completa
- `Part`: cada voz / instrumento
- `Measure`: compases (clave para an√°lise)


```python
parts = score.parts
print("N¬∫ de partes:", len(parts))
print("Nomes das partes:", [p.partName for p in parts])

# Primeiro comp√°s da primeira voz
m1 = parts[0].measure(1)
m1.show("text")
```

## 3) Extraer informaci√≥n b√°sica (comp√°s, tonalidade, tempo)

‚ö†Ô∏è Importante: `analyze('key')` √© unha estimaci√≥n.
En Harmon√≠a interesa ver se coincide co que ti analizar√≠as.


```python
k = score.analyze("key")
print("Tonalidade estimada:", k)

# Comp√°s: buscamos o primeiro TimeSignature que apareza
ts = score.recurse().getElementsByClass("TimeSignature").first()
print("Comp√°s:", ts)

# Tempo (pode non existir en obras antigas do corpus)
mm = score.recurse().getElementsByClass("MetronomeMark").first()
print("Tempo (se existe):", mm)
```

## 4) Exportar a MusicXML (para MuseScore)

1) Executa a cela  
2) Descarga o ficheiro `.musicxml` desde Files (Colab)  
3) Abre en MuseScore e exporta a PDF se queres.


```python
xml_path = score.write("musicxml", fp="S03_bwv66_6.musicxml")
xml_path
```
