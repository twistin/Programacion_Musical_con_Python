## 2) A nosa primeira partitura con `Stream`

Un `Stream` √© como un **contenedor temporal**:
- vai gardando eventos musicais en orde
- coma un pentagrama (ou unha parte) na partitura


```python
s = stream.Stream()

# Configuraci√≥n musical b√°sica
s.append(meter.TimeSignature("4/4"))
s.append(tempo.MetronomeMark(number=100))
s.append(key.Key("C"))  # Do maior

# Melod√≠a exemplo
melodia = ["C4", "D4", "E4", "G4", "E4", "D4", "C4"]
duracions = [1, 1, 1, 2, 1, 1, 2]

for p, d in zip(melodia, duracions):
    n = note.Note(p, quarterLength=d)
    s.append(n)

s.show("text")
```

### Exercicio r√°pido

## 3) Silencios (Rest) e acordes (Chord)

- `Rest` = silencio con duraci√≥n
- `Chord` = varias notas ao mesmo tempo


```python
s3 = stream.Stream()
s3.append(meter.TimeSignature("4/4"))
s3.append(tempo.MetronomeMark(number=90))

# 1 comp√°s de exemplo: acorde + silencio + nota longa
s3.append(chord.Chord(["C4", "E4", "G4"], quarterLength=2))
s3.append(note.Rest(quarterLength=1))
s3.append(note.Note("D4", quarterLength=1))

s3.show("text")
```
