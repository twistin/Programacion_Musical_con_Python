### Exercicio guiado
- Crea unha funci√≥n `repetir(mel, veces)`

```python
def repetir(mel, veces):
    resultado = []
    for _ in range(veces):
        resultado += mel
    return resultado

print(repetir(melodia, 2))
```

## 4) Clases: notas como obxectos

```python
class Nota:
    def __init__(self, pitch, dur):
        self.pitch = pitch
        self.dur = dur

    def __repr__(self):
        return f"Nota({self.pitch}, {self.dur})"

frase = [
    Nota("C4", 1),
    Nota("D4", 1),
    Nota("E4", 2)
]

frase
```

```python
# ‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
# üß™ AUTO-TEST: Verifica clase Nota
# ‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
assert 'Nota' in dir(), "‚ùå Debes crear a clase 'Nota'"

# Test: atributos correctos
n_test = Nota("C4", 1.0)
assert hasattr(n_test, 'pitch'), "‚ùå A clase Nota debe ter atributo 'pitch'"
assert hasattr(n_test, 'dur'), "‚ùå A clase Nota debe ter atributo 'dur'"
assert n_test.pitch == "C4", "‚ùå O pitch non se gardou correctamente"
assert n_test.dur == 1.0, "‚ùå A duraci√≥n non se gardou correctamente"

# Test: frase creada
assert 'frase' in dir(), "‚ùå Debes crear a variable 'frase'"
assert len(frase) >= 3, "‚ùå 'frase' debe ter polo menos 3 notas"

print("‚úÖ Clase Nota e frase correctas!")
```

### Transformaci√≥n sobre obxectos

```python
def alongar(frase, factor):
    return [Nota(n.pitch, n.dur * factor) for n in frase]

alongar(frase, 2)
```

## 5) Canon simple con desfase

Un canon √© unha repetici√≥n con atraso.

```python
def canon(frase, desfase):
    silencio = [Nota("REST", 1) for _ in range(desfase)]
    return frase + silencio + frase

mi_canon = canon(frase, 4)
mi_canon
```
